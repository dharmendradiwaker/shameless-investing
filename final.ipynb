##### Applying Nlp to make sure both files contain companies name in same formatting

```python
import pandas as pd
import re
from nltk.corpus import stopwords
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords



def clean_text(text):
    # Convert everything to a string, regardless of its original type
    text = str(text)
    
    # Remove "JOHNSON CONTROLS" from the text
    text = text.replace("Ltd.", "")
    
    # Find the index of the first "-" and trim everything after the last space before it
    dash_index = text.find("-")
    if dash_index != -1:
        last_space_before_dash = text.rfind(" ", 0, dash_index)
        text = text[:last_space_before_dash] if last_space_before_dash != -1 else text[:dash_index]
    
    # Remove punctuation
    text = re.sub(r'[^\w\s]', '', text)

    # Remove stopwords
    stop_words = set(stopwords.words('english'))
    text = ' '.join([word for word in text.split() if word not in stop_words])
    
    return text


# Apply the cleaning function to the 'Branch' column in both DataFrames
company_df['Name'] = company_df['Name'].apply(clean_text)

invest_df['stocks'] = invest_df['stocks'].apply(clean_text)


invest_df.drop(invest_df.columns[7:], axis=1,inplace=True)
invest_df.head(2)
```

```python
company_df.rename(columns={'Name': 'stocks'}, inplace=True)
company_df.head(1)
```

##### Marging both files

```python
invest_df['stocks_name'] = invest_df['stocks'].str.split().apply(lambda x: x[:2])
company_df['stock_Name'] = company_df['stocks'].str.split().apply(lambda x: x[:2])



# Apply matching function to filter rows for merging
merged_df = pd.merge(invest_df, company_df, how='outer')
print(merged_df.columns)

# Select desired columns for the output
output = merged_df[['Name', 'stocks', 'HOLDING VALUE', 'RUPEE', 'QTY HELD ', 'MAR24 CHANGE',
       'MAR24 HOLDING', 'S.No.', 'CMP Rs.',
       'Mar Cap Rs.Cr.', 'Div Yld %', 'NP Qtr Rs.Cr.', 'Qtr Profit Var %',
       'Sales Qtr Rs.Cr.', 'Qtr Sales Var %', 'ROCE %', 'Debt Rs.Cr.', 'P/E',
       'Earnings Yield %', 'EV / EBITDA', 'EV Rs.Cr.', 'PAT 12M Rs.Cr.',
       'Debt / Eq', 'Prom. Hold. %']]

output.to_excel('output.xlsx', index=False)
```
